# 2016/17

## Gameweeks

```python
element_gameweek_1617_df_arr = []

for i in range(1, 39):
    file_name = f'./data/1617/gws/gws/gw{i}.csv'
    element_gameweek_1617_df_arr.append(pd.read_csv(file_name, encoding='ISO-8859-1'))
    
element_gameweek_1617_df = pd.concat(element_gameweek_1617_df_arr)

element_gameweek_1617_df['event'] = element_gameweek_1617_df['round']

element_gameweek_1617_df[cols].to_csv('element_gameweeks_1617.csv', index=False)
```

```python
utils.write_to_table(
    'fpl',
    'element_gameweeks_1617',
    element_gameweek_1617_df[cols],
    secrets_path='../secrets/service_account.json',
    csvs_path='../csvs/'
)
```

## Elements

```python
element_1617_df = pd.read_csv('./data/1617/players_raw.csv')
```

```python
element_1617_df['element'] = element_1617_df['id']
element_1617_df['safe_web_name'] = element_1617_df['web_name'].apply(utils.get_safe_web_name)
element_1617_df['safe_team_name'] = element_1617_df['team'].apply(lambda x: teams_1617[x])

element_1617_df = element_1617_df[[
    'element',
    'safe_web_name',
    'element_type',
    'team',
    'safe_team_name',
    'first_name',
    'second_name'
]]

element_1617_df.head()
```

```python
utils.write_to_table(
    'fpl',
    'elements_1617',
    element_1617_df,
    secrets_path='../secrets/service_account.json',
    csvs_path='../csvs/',
    write_disposition='WRITE_TRUNCATE'
)
```

## Previous seasons

```python
previous_season_1617_df_arr = []

for i in os.listdir('./data/1617/players/'):
    first_name = i.split('_')[0]
    second_name = i.split('_')[1]
    try:
        df = pd.read_csv(f'./data/1617/players/{i}/history.csv')
        df['first_name'] = first_name
        df['second_name'] = second_name
        previous_season_1617_df_arr.append(df)
    except:
        print(first_name, second_name)
```

```python
previous_season_1617_df = pd.concat(previous_season_1617_df_arr)

previous_season_1617_df = previous_season_1617_df.join(
    element_1617_df.set_index(['first_name', 'second_name']),
    on=['first_name', 'second_name'],
    how='inner'
)

previous_season_1617_df = previous_season_1617_df[previous_season_cols]
previous_season_1617_df.head()
```

```python
utils.write_to_table(
    'fpl',
    'element_previous_seasons_1617',
    previous_season_1617_df,
    secrets_path='../secrets/service_account.json',
    csvs_path='../csvs/',
    write_disposition='WRITE_TRUNCATE'
)
```
